```python
import pandas as pd
import statsmodels.api as sm
```

```python
train = 'D:/estrategia_/train.csv'

df = pd.read_csv(train)
df
```

```python
y = 'SalePrice'
columns = ['MSSubClass', 'LotFrontage', 'LotArea', 'YrSold', 'SalePrice']
df = df[columns]
df = df.dropna()
df
```

```python
X = df.drop([y], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
normalized_x = scaler.fit_transform(X)
normalized_x = pd.DataFrame(normalized_x, columns=X.columns)
```

```python
modelo = sm.OLS(df[y].values, normalized_x)
```

```python
res = modelo.fit()
```

```python
res.summary()
```
