```python
import pandas as pd
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
df =  pd.read_excel('full_grouped_filtro.xlsx')
```

```python
df.head(6)
```

```python
df = df.replace("Argentina", "1")
df = df.replace("Chile", "2")
df = df.replace("Ecuador", "3")
df = df.replace("Mexico", "4")
df = df.replace("Spain", "5")
df = df.replace("Americas", "1")
df = df.replace("Europe", "2")
df.head()
```

```python
df.info()
```

```python
'''df.Date = df.Date.astype('category')
df.Date = df.Date.cat.codes'''
```

```python
df.head()
```

```python
X = df.drop(['New deaths','Date'], axis=1)
y = df['New deaths']
X_std = StandardScaler().fit_transform(X)

# Separar o dado em conjunto de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=10)

# Modelo
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

# Coeficiente de determinacao
print(r2_score(y_test, y_pred))

# Erro medio absoluto
print(mean_absolute_error(y_test, y_pred))
```
